# Variables categóricas

```python
categorical=df.select_dtypes(include=object).columns
variables=categorical.drop('class')
for c in variables:
  print(c)
  print(df[c].unique())
```

Se extrae el género de la columna personal_status porque no hay una similitud en los estados civiles disponibles en el dataset entre hombres y mujeres

```python
df['sex']= df['personal_status'].str.extract(r'\b(male|female)\b')
df.drop(['personal_status'], axis=1, inplace = True)
```

Se eliminan los simbolos de >, < y = de las columnas ***checking_status*** ,***saving_status*** y ***employment*** para darle una mayor legibilidad a la base de datos y a los gráficos.

```python
df['checking_status'].replace(['<0' ,'0<=X<200' , '>=200'],['less than 0','0 to 199','200 or more'], inplace=True)
df['savings_status'].replace(['<100', '500<=X<1000', '>=1000' ,'100<=X<500'],['less than 100','500 to 999', '100 to 499','1000 or more'], inplace=True)
df['employment'].replace(['>=7', '1<=X<4', '4<=X<7', '<1'],['7 or more','1 to 3', '4 to 6','less than 1'], inplace=True)
```

```python
df.describe()
```

```python
categorical=df.select_dtypes(include=object).columns
variables=categorical.drop('class')
categoric_titles=['Estado de la cuenta corriente','Historial crediticio del usuario','Propósito del crédito','Estado de la cuenta de ahorro',
                  'Años de empleo', 'Otros deudores', 'Propiedades','Otros planes de pago','Tipo de vivienda', 'Tipo de trabajo',
                  'Propietario de teléfono','Trabajador extranjero','Sexo']
```

```python
num_plots = len(variables)
num_cols = 2
num_rows = (num_plots + 1) // num_cols

fig, axes = plt.subplots(num_rows, num_cols, figsize=(12, 30))


for i, variable in enumerate(variables):
    row = i // num_cols
    col = i % num_cols
    ax = axes[row, col]
    sns.countplot(data=df, x=variable, ax=ax)
    ax.set_title(f'Countplot of {variable}', fontsize=14)
    ax.set_xlabel(variable, fontsize=12)
    ax.set_ylabel('Count', fontsize=12)
    ax.tick_params(axis='x', rotation=45)

if num_plots < num_rows * num_cols:
    for i in range(num_plots, num_rows * num_cols):
        fig.delaxes(axes.flatten()[i])


plt.tight_layout()


plt.show()
```
